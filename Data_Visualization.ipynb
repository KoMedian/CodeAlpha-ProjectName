```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("posts.csv")
```

```python
x = df ['Post_Type'].values
y = df ['likes'].values
plt.scatter(x,y,color='lightgreen')
plt.xlabel('Post Type -->')
plt.ylabel('Likes -->')
plt.title('Linear Regression')
df.describe
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
import numpy as np

le = LabelEncoder()
x_encoded = le.fit_transform(x.ravel()).reshape(-1, 1)
reg = LinearRegression()
reg.fit(x_encoded, y)

# Predict
y_pred = reg.predict(x_encoded)

print("Predictions:", y_pred)
print("Slope:", reg.coef_)
print("Intercept:", reg.intercept_)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x_flat = x.flatten()
y_flat = y.flatten() if y.ndim > 1 else y
plt.scatter(x_flat, y_flat, color='red', label='Data Points')
sorted_indices = np.argsort(x_flat)
plt.plot(x_flat[sorted_indices], y_pred[sorted_indices], color='grey', label='Regression Line')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()

# Show plot
plt.show()
```
